# Unit 3 AP CSP Key Takeaways

## 3.1 Variables & Assignments
- Variables are like containers that hold information
- You can change what's stored in a variable
- Use the symbol to put a value into a variable
- Give variables names that make sense so you remember what they're for
- Example: age 15 means the variable "age" now holds 15

## 3.2 Data Abstraction
- Lists let you store a bunch of things in one place instead of making tons of separate variables
- Makes it way easier to organize data
- You can access items in a list by their position number
- Lists are just a simpler way to handle lots of information at once

## 3.3 Mathematical Expressions
- You can do math in code just like regular math (adding, subtracting, etc.)
- MOD gives you the remainder when you divide
- Order of operations still matters (multiply/divide before add/subtract)
- The computer calculates everything and gives you one answer

## 3.4 Strings
- Strings are just text in quotes
- You can combine strings together (like "hello" + "world")
- You can grab part of a string if you need to
- Each letter in a string has a position number

## 3.5 Boolean Expressions
- These are statements that are either TRUE or FALSE
- You use them to compare things (is this equal to that? is this bigger?)
- AND, OR, and NOT help you combine conditions
- These are super important for making decisions in code

## 3.6 Conditionals
- IF statements let your program make choices
- "IF this is true, do this thing"
- IF-ELSE gives you a backup plan if the condition is false
- Only one part of the IF-ELSE will actually run

## 3.7 Nested Conditionals
- This is basically putting an IF statement inside another IF statement
- Lets you check multiple things in order
- Can get confusing if you do too many layers
- Helps make more specific decisions

## 3.8 Iteration
- Iteration is just a fancy word for repeating code (loops)
- REPEAT lets you do something multiple times without writing it over and over
- REPEAT UNTIL keeps going until something becomes true
- Be careful not to make loops that never stop (infinite loops)

## 3.9 Developing Algorithms
- An algorithm is just the steps to solve a problem
- Break big problems into smaller chunks
- You use sequences (order), decisions (IF), and loops (REPEAT)
- Test your algorithm to make sure it actually works

## 3.10 Lists
- Lists can change size - you can add or remove items
- You can add things to the end, insert in the middle, or remove stuff
- Going through each item in a list is called traversing
- Lists make it easy to work with a bunch of related data

## 3.12 Calling Procedures
- Procedures are chunks of code you can reuse by calling their name
- Instead of writing the same code over and over, just call the procedure
- Procedures can take inputs (parameters) to work with different values
- Makes your code cleaner and easier to read

## 3.13 Developing Procedures
- When you make a procedure, you give it a name and tell it what to do
- Parameters are the inputs you give the procedure
- Procedures can send back an answer (return value)
- Each procedure should do one specific job
- Writing what the procedure does helps you remember later

## 3.15 Random Values
- RANDOM gives you an unpredictable number in a range you choose
- Good for games, simulations, or when you need something random
- The computer doesn't actually pick truly random numbers, but close enough
- Useful when you want variety or to test things

## 3.17 Algorithmic Efficiency
- Some ways of solving a problem are faster than others
- Efficiency is about how much time or memory your code uses
- When you have more data, efficiency matters more
- Sometimes you compare different solutions to see which is better
- Doesn't have to be perfect, just understanding that some methods are quicker

# Trimester Retrospective: My Growth This Year

## Looking Back: How I've Changed

At the start of the year, I honestly didn't know much about computer science. I could follow tutorials and copy code examples, but I didn't really understand why things worked. If something broke, I'd be pretty lost. Now, things are different. I can actually debug problems on my own, I understand what's happening behind the scenes, and I can explain technical concepts without stumbling over my words. It's wild to look back and see how much I've learned.

## Sprint Reflections

### Sprint 1: Tools and Environment
Learning about tools like Git, development environments, and deployment was kind of boring at first, not gonna lie. I just wanted to code! But now I get why it matters. Knowing how to use version control properly, set up my workspace, and manage dependencies makes everything so much easier. I used to think these were just extra steps, but now I realize they're what separate messy projects from professional ones.

### Sprint 2: Fundamentals of JavaScript/Python
This sprint was where things started clicking. Learning JavaScript and Python side-by-side helped me see patterns that work across different languages. I finally understood things like how loops really work, what data structures to use when, and how to work with APIs. There were definitely moments where I felt confused, but then something would click and I'd be like "ohh, that's how that works!" Those moments kept me going.

### Sprint 3: West Coast Travel / Digital Famine
This project was the hardest but also the most fun interesting part. Taking everything I learned and actually building something real was tough.
